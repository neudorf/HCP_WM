````python
# !pip install pandas openpyxl

import pandas as pd
from pathlib import Path

# =====================================================
# CHANGE ONLY THIS PATH
# =====================================================
BASE_PATH = Path("/content/drive/Shareddrives/Impact Scholars (Scholars External Drive)/DataSets/HCP_WM_Unrelated-100-subjects/data/SupplementaryTables")

OUTPUT_MD = BASE_PATH / "supplementary_material.md"

# =====================================================
# INPUT FILES
# =====================================================
table_files = [
    {
        "file": "unique_connections_top50_face.xlsx",
        "title": "Supplementary Table S1",
        "name": "tbl-face-connections",
        "caption": "Top 50 unique functional network connections for faces."
    },
    {
        "file": "unique_connections_top50_body.xlsx",
        "title": "Supplementary Table S2",
        "name": "tbl-body-connections",
        "caption": "Top 50 unique functional network connections for bodies."
    },
    {
        "file": "unique_connections_top50_places.xlsx",
        "title": "Supplementary Table S3",
        "name": "tbl-places-connections",
        "caption": "Top 50 unique functional network connections for places."
    },
    {
        "file": "unique_connections_top50_tools.xlsx",
        "title": "Supplementary Table S4",
        "name": "tbl-tools-connections",
        "caption": "Top 50 unique functional network connections for tools."
    },
    {
        "file": "common_connections_top50.xlsx",
        "title": "Supplementary Table S5",
        "name": "tbl-shared-connections",
        "caption": "Top 50 shared functional network connections."
    },
    {
        "file": "panel_F_node_labels_short_full.csv",
        "title": "Supplementary Table S6",
        "name": "tbl-node-labels",
        "caption": "Node labels and network assignments."
    }
]

# =====================================================
# CONVERT DATAFRAME TO MyST LIST-TABLE
# =====================================================
def df_to_myst_list_table(df, title, name, caption):
    columns = list(df.columns)

    lines = [
        f"## {title}",
        "",
        f"```{{list-table}} {caption}",
        ":header-rows: 1",
        f":name: {name}",
        "",
        "* - " + "\n  - ".join(columns),
        ""
    ]

    for _, row in df.iterrows():
        values = []

        for val in row:
            if pd.isna(val):
                values.append("")
            elif isinstance(val, float):
                values.append(f"{val:.3f}")
            else:
                values.append(str(val))

        lines.append("* - " + "\n  - ".join(values))
        lines.append("")

    lines.append("```")
    return "\n".join(lines)

# =====================================================
# READ FILES AND GENERATE MARKDOWN
# =====================================================
output = ["# Supplementary Material", ""]

for item in table_files:
    file_path = BASE_PATH / item["file"]

    print(f"Loading: {file_path}")

    if file_path.suffix.lower() == ".csv":
        df = pd.read_csv(file_path)
    else:
        df = pd.read_excel(file_path)

    output.append(
        df_to_myst_list_table(
            df=df,
            title=item["title"],
            name=item["name"],
            caption=item["caption"]
        )
    )

    output.append("")

# =====================================================
# SAVE MARKDOWN
# =====================================================
with open(OUTPUT_MD, "w", encoding="utf-8") as f:
    f.write("\n".join(output))

print(f"\nSaved markdown file:\n{OUTPUT_MD}")
````
