## Functions:

```python
def get_patient_numbers(numbers_string, debug):
    """
    Returns a NumPy array of patient numbers from a string.

    numbers_string - is a string that can be generated my running ls in console
    debug - is a bool that gives more printouts and checks
    Set to false for better preformance
    """
    # Split the string by whitespace and convert to a list of integers
    patient_numbers_list = [int(num) for num in numbers_string.split() if num.strip()]
    # Convert the list to a NumPy array
    patient_numbers_array = np.array(patient_numbers_list)
    if debug:
      print("NumPy array of patient numbers:")
      display(patient_numbers_array)
      print("Shape of the array:", patient_numbers_array.shape)

    return patient_numbers_array
```

```python
# We use Glassier parcellation from nilearn
# and apply it with Glassier Parcels
# by extracting parcel time series
# (mean signal per parcel of all voxels within that parcel/area) per time point


patient_number = "100307"
patient_number_string = str(patient_number)
if debug:
  print("Patient number:", patient_number)


dtseries_file_path = "/WM/LR/tfMRI_WM_LR_Atlas_MSMAll_hp0_clean_rclean_tclean.dtseries.nii"
dtseries_file = main_path + patient_number_string + dtseries_file_path

# Load dtseries file, (timepoints, grayordinates):
data = nib.load(dtseries_file)
if debug:
  print("Data shape :", data.shape)

# HCP Normlization:
data_norm = hcp.normalize(data.get_fdata())
if debug:
  print("Data norm shape:", data_norm.shape)

# HCP Parcellation:
data_norm_parcel = hcp.parcellate(data_norm, hcp.mmp)
if debug:
  print("Data norm parcel shape:", data_norm_parcel.shape)

# Get cortical grayordinates using CIFTI brain models
cifti_header = data.header

# Get brain models as a list
brain_models = list(cifti_header.get_index_map(1).brain_models)

# Identify left and right cortical grayordinates
lh_idx = np.arange(brain_models[0].index_count)  # left cortex
rh_idx = np.arange(brain_models[1].index_count) + brain_models[0].index_count  # right cortex
if debug:
  print("Left cortical grayordinates:", lh_idx, " Total shape:", lh_idx.shape)
  print("Right cortical grayordinates:", rh_idx, " Total shape:", rh_idx.shape)

# Combine cortical indices from left 0-29695 and right 29696-59411
cortical_idx = np.concatenate([lh_idx, rh_idx])
data_cortical = data_norm[:, cortical_idx]
if debug:
  # Matchs Glasser labels: (time, 59412)
  print("Cortical data shape:", data_cortical.shape)

```

```python
# 4D Data:
nii_path_vol = main_path+patient_number+"/WM/LR/tfMRI_WM_LR_hp0_clean_rclean_tclean.nii.gz"
# A 4D NIFTI volume: Stored in subject (native) volumetric space - Native volume (not surface-based).
# One value per voxel, per time point.
# Hemisphere info: Implicit in volume (not split)
# KEYS: hp0 → no high-pass filtering, clean → ICA-FIX denoising applied,
# rclean → spatially cleaned, tclean → temporally cleaned

# Load NIfTI:
img_vol = nib.load(nii_path_vol)

# 2D Data and only cortical data:
nii_path_surf = main_path+patient_number+"/WM/LR/tfMRI_WM_LR_Atlas_MSMAll_hp0_clean_rclean_tclean.dtseries.nii"
# Combines surface vertices + subcortical gray matter voxels
# Space: HCP grayordinate space. Cortical data on the fs_LR 32k surface. Subcortical data in standard atlas space.
# Hemisphere info: Explicit (left/right surfaces)
# KEYS: Atlas → mapped to standard grayordinate atlas.
# MSMAll → aligned using Multimodal Surface Matching (uses folding, myelin, functional features)
img_surf = nib.load(nii_path_surf)

# Binary Mask in brain or out of brain:
nii_path_mask = main_path+patient_number+"/WM/RL/brainmask_fs.2.nii.gz"
# A 3D volumetric brain mask. Binary (or near-binary): brain = 1, non-brain = 0
# KEYS: “fs” → derived from FreeSurfer segmentation. “.2” → HCP versioning
img_mask = nib.load(nii_path_mask)
#-------------------------------------------------------------------------------

# Inspect 4D data:
print("File:", nii_path_vol)
print("Shape:", img_vol.shape)                  # (X, Y, Z, T) = 91, 109, 91, 405
print("Voxel sizes / TR:", img_vol.header.get_zooms())
print("Data type:", img_vol.get_data_dtype())
print("Units:", img_vol.header.get_xyzt_units())
print("Affine:\n", img_vol.affine)
# X=left-to-right, Y=posterior-to-anterior, Z=inferior-to-superior
data_vol = img_vol.get_fdata() # to NumPy array

# Inspect 2D data:
print("File:", nii_path_surf)
print("Shape:", img_surf.shape)                  # (T, grayordinates) = 405, 91282
print("Data type:", img_surf.get_data_dtype())
data_surf = img_surf.get_fdata() # to NumPy array
# 64984 = 32492 × 2 (cortex: 2 hemisphere) + 26298 (subcortex) = 91282
# 29696 + 29716 = cortex in data

# Inspect Binary Mask:
print("File:", nii_path_mask)
print("Shape:", img_mask.shape)                  # (X, Y, Z) = (91, 109, 91)
print("Data type:", img_mask.get_data_dtype())
data_mask = img_mask.get_fdata() # to NumPy array

#-------------------------------------------------------------------------------

count_true = np.count_nonzero(data_mask)
count_false = data_mask.size - count_true
print(np.mean(data_mask)) # 25% brain, 75% outside brain

# we need to use this 4d data for each subject, and do Glassier parcellation:
# 32,492 × 2 -> 180 x 2, and ignoring subcortex

# The mapping is stored in the CIFTI header, not in the data array.
hdr = img_surf.header
brain_models = hdr.get_index_map(1).brain_models
for bm in brain_models:
    print(bm.brain_structure, bm.index_count)
```

```python
# @title

# ------------------------------------------------------------------------------
# Load Glasser 360 parcellation from a github, need to double check it is the correct file:
# TODO.
parc_file = "/content/drive/Shareddrives/Impact Scholars (Scholars External Drive)/DataSets/Glasser360.32k_fs_LR.dlabel.nii"
parc_img = nib.load(parc_file)
# Extract label per grayordinate
labels = parc_img.get_fdata().squeeze()  # shape: (grayordinates,) = (59412,)
print("Labels shape:", labels.shape)

data_norm_parcel = hcp.parcellate(data_norm, hcp.mmp) # another name: mmp = Glasser; 360 + 19 (sub cortical layers)
print("Labels from HCP shape:", data_norm_parcel.shape)

# Average per parcel
unique_parcels = np.unique(labels[labels>0])
unique_parcels.sort()
parcel_timeseries = np.zeros((data_norm.shape[0], len(unique_parcels)))

for i, p in enumerate(unique_parcels):
    parcel_mask = labels == p
    parcel_timeseries[:, i] = data_cortical[:, parcel_mask].mean(axis=1)

print("Parcel timeseries shape:", parcel_timeseries.shape)  # (time, 360)

#time_point = 30
#instant_parcel_data = parcel_timeseries[time_point, :]
#plotting.view_surf(hcp.mesh.inflated, hcp.cortex_data(hcp.unparcellate(instant_parcel_data[:], hcp.mmp), fill=0), threshold=0.1, bg_map=hcp.mesh.sulc)
```

```python
# Your data
# instant_parcel_data shape: (405, 360)
timepoint = 0  # change this to any timepoint 0-404
activation = parcel_timeseries[timepoint, :]  # shape: (360,)

# Load Glasser atlas
atlas = nib.load(parc_file)
parcel_labels = np.array(atlas.get_fdata()).squeeze()  # shape: (59412,)

# Create an array for the 59412 cortical grayordinates to store activation values
cortical_activation_grayordinates = np.zeros(parcel_labels.shape[0])

# Map 360 parcel values to the 59412 cortical grayordinates
for i in range(360):
    cortical_activation_grayordinates[parcel_labels == (i + 1)] = activation[i]

# Create a full 91282 grayordinate array and fill the cortical parts
# cortical_idx is available from previous cell (7pVvZyrtcT3x)
full_grayordinate_data = np.zeros(91282)
full_grayordinate_data[cortical_idx] = cortical_activation_grayordinates

# Use hcp.cortex_data to map the full grayordinate data to the 64984 cortical vertices
# This correctly handles the mapping to the fs_LR 32k surface for plotting.
vertex_data = hcp.cortex_data(full_grayordinate_data, fill=0) # shape (64984,)

# Split hemispheres into 32492 vertices each for fs_LR 32k surface
lh_vertex = vertex_data[:32492]
rh_vertex = vertex_data[32492:]
from neuromaps.datasets import fetch_fslr
from surfplot import Plot

# Get surfaces
surfaces = fetch_fslr()
lh_surf = surfaces['inflated'][0]
rh_surf = surfaces['inflated'][1]

# Plot
p = Plot(lh_surf, rh_surf, views=['lateral', 'medial'])
p.add_layer(
    {'left': lh_vertex, 'right': rh_vertex},
    cmap='RdBu_r',
    color_range=(-activation.std()*3, activation.std()*3)
)
fig = p.build()
fig.savefig(f'activation_t{timepoint}.png', dpi=300)
plt.show()
```
